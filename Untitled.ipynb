```python
import scipy
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt
plt.ion()
```

```python
data_normal = scipy.stats.norm.rvs(size = 10000, loc=0, scale=1) # mean and starndard deviation.
```

```python
data_uniform = scipy.stats.uniform.rvs(size= 10000, loc=3, scale=5) # Has start and width.
```

```python
data_gamma = scipy.stats.gamma.rvs(size=10000, loc=0, a=1) # shape and rate parameter. erlang distrubtion
```

```python
dists = pd.DataFrame({'norm':data_normal, 'uni':data_uniform, 'gamma':data_gamma}, columns=['norm', 'uni', 'gamma'])
```

```python
dists.norm.plot()
```

```python
dists.uni.plot()
```

```python
dists.gamma.plot()
```
