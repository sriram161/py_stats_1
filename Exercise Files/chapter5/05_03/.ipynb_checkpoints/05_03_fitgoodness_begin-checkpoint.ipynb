Loading gapminder data for year 1985 (Live Aid!) and setting up plot as in chapter 3

```python
gapminder = pd.read_csv('gapminder.csv')
```

```python
gdata = gapminder.query('year == 1985')
```

```python
size = 1e-6 * gdata.population

colors = gdata.region.map({'Africa': 'skyblue', 'Europe': 'gold', 'America': 'palegreen', 'Asia': 'coral'})

def plotdata():
    gdata.plot.scatter('age5_surviving','babies_per_woman',
                      c=colors,s=size,linewidths=0.5,edgecolor='k',alpha=0.5)
```

Setting up model plot

```python
def plotfit(fit):
    plotdata()
    pp.scatter(gdata.age5_surviving,fit.predict(gdata),
               c=colors,s=30,linewidths=0.5,edgecolor='k',marker='D')
```

Three models from last video

```python
groupmeans = smf.ols(formula='babies_per_woman ~ -1 + region', data=gdata).fit()
```

```python
surviving = smf.ols(formula='babies_per_woman ~ -1 + region + age5_surviving', data=gdata).fit()
```

```python
surviving_byregion_population = smf.ols(
    formula='babies_per_woman ~ -1 + region + age5_surviving'
            '+ age5_surviving:region - age5_surviving + population',
    data=gdata).fit()
```

```python
plotfit(surviving)
```

```python
plotfit(surviving_byregion_population)
```

```python
## Which model is better??
#### From the above we plot 2 models surviving and surviving_byregion_population. Here we look at classical measures of goodness of fit.
```

### Classical models of goodness of fit such as 
1. mean square error (MSE)
2. F-score
3. R-squared
4. ANOVA table
#### The above method are developed in pre-computer era analytics heavy needs little mathematical background to understand.

```python
# We can Do scatterplot of a explanatory variable aganist he residuals
```

```python
pp.scatter(gdata.age5_surviving, groupmeans.resid) # resid give residual errors.
```

```python
# Comparing plots of 3 model residuals.
pp.scatter(gdata.age5_surviving, groupmeans.resid)
pp.scatter(gdata.age5_surviving, surviving.resid)
pp.scatter(gdata.age5_surviving, surviving_byregion_population.resid)
pp.legend(['groupmean', 'surviving', 'surviving_byregion_population'])
```

```python
# [Observation] From the above plot surviving_byregion_population performs better so for than simpler models.
```

```python
# [NOTE] : mean square error is the simplest error to look at.
for model in (groupmeans, surviving, surviving_byregion_population):
    print(model.mse_resid)
```

# R-squared
1. R_squared = (Explained Variance)[variance of of the fitted model] / (Total Variance)
#### R_squared range [0, 1]
#### R_squared tends to 1 means model is better performing.

```python
for model in (groupmeans, surviving, surviving_byregion_population):
    print(model.rsquared)
```

# F-statistic
#### Whcih takes into account not only explained variance but also the number of model parameters.
#### From F-statistics we gain insight that too many parameter will overfit the data by losing predictive power.

##### John von Neumann quote "With four parameters I can fit an elephant but with five I can wiggle his trunk"

# The F-statistic measures how much on an average each parameter contributes to the growth of the R-squared. compared to a rondom model vector(hypotherical explanatory variable that has random values and no link to response variable)

### F is 1 parameter explains same as random vector
### F is >> 1 means parameter expalins more than random vector.

```python
for model in (groupmeans, surviving, surviving_byregion_population):
    print(model.fvalue)
```

###### [Observation] From the above F-score we can observe that surviving model [150.xxxx] explains better than other models.

```python
surviving.summary()
```

# ANOVA
### Certain key goodness of fit are collected in ANOVA table
1. df -> No of degrees of freedom for regions is number of parameters for Residuals number of datapoints - no.of parameters.
2. sum_sq -> total  variance
3. mean_sq -> mean of variance
4. F -> F-statistic
5. -> P-value for a known models with same parameter as ours where all terms are random.

```python
sm.stats.anova_lm(groupmeans)
```

```python
sm.stats.anova_lm(surviving)
```

```python
sm.stats.anova_lm(surviving_byregion_population)
```
