```python
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf

# fix error in some versions of statsmodels
import scipy
if not hasattr(scipy.stats,'chisqprob'):
    scipy.stats.chisqprob = lambda chisq, df: scipy.stats.chi2.sf(chisq, df)
```

```python
smoking = pd.read_csv("whickham.csv")
```

```python
smoking = smoking[smoking.age < 65]
```

```python
color = np.where(smoking.smoker == 'No','skyblue','coral')
jitter = 0.05*(np.random.rand(len(smoking)) - 0.5)
def plotfit(fit):
    pp.scatter(smoking.age,smoking.outcome + jitter,c=color,linewidth=0.2,edgecolor='k',alpha=0.5)
    pp.scatter(smoking.age,fit.predict(smoking),c=color,linewidth=0.2,edgecolor='gray',marker='d')
```

```python
# Multiplying a Bool by an integer returns an integer
smoking['outcome'] = 1*(smoking['outcome'] == 'Alive')
```

```python
ols_group_age = smf.ols(formula='outcome ~ smoker + age',data=smoking).fit()
```

```python
plotfit(ols_group_age)
```

```python
# The data is plotted as circles orange -> Smokers, Lightblue -> Non-Smokers 
# Fit is represented by diamonds.
# We see an assocation between smoking and negative outcomes.
# The smoker term is -ve.
# The above plot shows a problem some predicted outcomes are larger than one. <- Problem.
```

```python
sm.stats.anova_lm(ols_group_age)
```

```python
ols_group_age.conf_int() # Return confidence intervals.
```

```python
# statsmodel implements it as logit
```

```python
logit_smoker_age = smf.logit(formula = 'outcome ~ smoker + age', data = smoking).fit()
```

```python
plotfit(logit_smoker_age)
```

```python
# Model is bounded between 0 and 1.
# Displays a non-linear behaviour even if all linear main terms.
# The criteria for Logistic regression is not minimizing mean square error.
# Intrepeting the model response as a propability function and maximizing the resulting probability of the observed data this 
# is a form of maximum likelyhood estimation.
```

```python
logit_smoker_age.llf
```

```python
logit_smoker_age.params
```

```python
# Smoking reduce the probability of being alive but not uniformly.
```

```python
logit_smoker_age.conf_int() # association with smoking remains weak.
```

```python
logit_smoker_age.summary()
```
