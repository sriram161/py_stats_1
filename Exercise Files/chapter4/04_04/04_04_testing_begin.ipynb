#### This exercise test how to test a null hypothesis using the Dr. Jhon Snow's Cholora and water pummp data analysis.

#### Hypothesis testing Steps:
* Make an observation 
* Define a hypothesis
 *  After you have a question, to start you should have a answer with limited evidence. i.e There is no relation between the causing variable(you think) and effect.
* Choose a statistic
* Select a p-value
*  Simulate the sampling distribution for the observation.
* Check the quatile of the observation in sampling distribution.
* Accept or reject the null-hypothesis.

#### Process we falled in this example
1. We have made an observation (Many deaths in area 0 due to Cholera)
2. We have made a hypothesis (It is Pump has nothing to do with cholera)
3. estimate a sampling distribution of expected deaths under a null hypothesis ( Deaths are caused near null pumps.)
4. Observe how extreme our observed finding was with respect to null distribution.. (Find the p-value of the observation by checking in which quantile the observation falls.)
5. Conclude by rejection of null-hypothesis or fail to reject (In the example We Reject the null- hypothesis cause the p-value is small.)

####[Observation] IF deaths occured randomly in each area theres is no way we could get 340 deaths in area zero.

```python
# Here we assume population living in each area are very different and many more people leaving at pump zero. We expect more deaths there.
```

```python
# We are simulationg the deaths to form a distribution
```

```python
def simulate(n):
    return pd.DataFrame({'closest': np.random.choice([0,1,4,5], size=n, p=[0.65, 0.15, 0.1, 0.1])})
```

```python
simulate(489).closest.value_counts()
```

```python
# Create a sampling distribution for number of deaths in area zero [stastics is count of death]
```

```python
area = 0
sampling = pd.DataFrame({'counts': [simulate(489).closest.value_counts()[area] for _ in range(10000)]})
```

```python
sampling.counts.hist(histtype='step')
```

```python
#### This sampling distribution is that pumps has nothing to do with deaths due to cholera at area zero.
#### [From Null-Hypothesis] Deaths occur simply proportional population size.
#### We expect the observed value fall near to the mean in accordance with null-hypothesis.
```

```python
scipy.stats.percentileofscore(sampling.counts, 340) # This methods give the percentile of value from the distribution.
```

```python
# p-value we expected.
100-98.72
```

```python
#### Since p-value is < 5 ie. 1.28 
#### [Conclusion] We reject the null-hypothesis
#### We verified how extreme our observation was in a null-hypothesis sampling distribution.
```
